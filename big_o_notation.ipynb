```python
plt.figure(figsize =(12, 6))

n = np.array(range(1, 11))

y1 = 0 * n + 1
y2 = np.log(n)
y3 = n
y4 = n * np.log(n)
y5 = n ** 2
y6 = 2 ** n
y7 = n * (n - 1)
y8 = n ** n

plt.subplot(111)
plt.plot(n, y1, c='blue')
plt.plot(n, y2, c='red')
plt.plot(n, y3, c='yellow')
plt.plot(n, y4, c='green')
plt.plot(n, y5, c='black')
plt.plot(n, y6, c='orange')
plt.plot(n, y7, c='purple')
plt.plot(n, y8, c='pink')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(['$O(1)$', 'O(log n)', '$O(n)$', 'O(n log n)', '$O(n^2)$', '$O(2^n)$', '$O(n!)$', '$O(n^n)$'])
plt.axis([0, 20, 0, 90])
```
